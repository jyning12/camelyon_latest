## 第一个，原始

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

train_log = open("log64100_01.log")
log_line = train_log.readlines()
epoch, acc, loss, val_acc, val_loss = [], [], [], [], []
for i in range(len(log_line)):
    if i<len(log_line)-1:
        line_list = list(eval(log_line[i+1]))
        epoch.append(line_list[0])
        acc.append(line_list[1])
        loss.append(line_list[2])
        val_acc.append(line_list[3])
        val_loss.append(line_list[4])
fig, axes = plt.subplots(2,2,figsize=(20,10))    # 自定义每个小图的尺寸
ax = axes.ravel()
ax[0].plot(epoch,acc)
ax[0].set_title('train acc')
# ax[0].set_ylim([0.0, 1.05])

ax[1].plot(epoch,loss)
ax[1].set_title('train loss')
# ax[1].set_ylim([0.0, 1.05])

ax[2].plot(epoch,val_acc)
ax[2].set_title('validation acc')
# ax[2].set_ylim([0.0, 1.05])

ax[3].plot(epoch,val_loss)
ax[3].set_title('validation loss')
# ax[3].set_ylim([0.0, 1.05])
```

    上面这个是epoch=100,batch_size=64,steps_per_epoch=100,SGD(lr=0.05,momentum=0.9,decay=0.5)
    保存文件为model64100_01.h5、log64100_01.log
